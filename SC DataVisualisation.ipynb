## **Bar** **Graph**

```
plt.bar?
```

```
plt.bar([0,1,2],[10,20,15])
plt.show()
```

```
# Bar Graph is overlapping
plt.bar([0,1,2],[10,20,15]) #Current Year
plt.bar([0,1,2],[20,10,25]) #Current Year
plt.show()
```

```
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates,[10, 20, 15], width = 0.5)
plt.bar(x_coordinates+0.5,[20,10,25], width = 0.5)
plt.show()
```

```
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates,[10, 20, 15], width = 0.5, label = "Current Year")
plt.bar(x_coordinates+0.5,[20,10,25], width = 0.5, label = "Next Year")
plt.legend()
plt.show()
```

```
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates-0.25,[10, 20, 15], width = 0.5, label = "Current Year", tick_label = ["Gold", "Silver", "Platinum"])
plt.bar(x_coordinates+0.25,[20, 10, 25], width = 0.5, label = "Next Year")
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```
plt.style.use("dark_background")
```

## **Pie Chart**

```
plt.pie?
```

```
subjects = "Machine Learning", "Business Comm.", "Compiler Design", "Graph Theory"
weightage = [20, 10, 30, 40]
```

```
plt.pie(weightage, labels = subjects)
plt.show()
```

```
# Highlight
plt.pie(weightage, labels = subjects, explode=(0,0,1,0))
plt.show()
plt.pie(weightage, labels = subjects, explode=(1,0,1,0))
plt.show()
```

```
plt.pie(weightage, labels = subjects, explode=(0,0,.1,0))
plt.show()
plt.pie(weightage, labels = subjects, explode=(.1,0,.1,0))
plt.show()
```

```
# Shows percentage - (autopct)
plt.pie(weightage, labels = subjects, explode=(0,0,.1,0), autopct='%1.1f%%')
plt.show()
plt.pie(weightage, labels = subjects, explode=(.1,0,.1,0), autopct='%1.1f%%')
plt.show()
```

```
plt.pie(weightage, labels = subjects, explode=(0,0,.1,0), autopct='%1.1f%%', shadow=True)
plt.show()
```

```
plt.style.use("seaborn")
```

# **Histogram**

```
xsn = np.random.randn?
```

```
xsn = np.random.randn(100)
print(xsn)
```

**sigma = r, u = x => It means that maximum the random numbers generated will lie between (x-r) to (x+r)**

```
Xsn = np.random.randn(100)
sigma = 5
u = 60
X = np.round(Xsn*sigma +u)
print(X)
```

```
Xsn = np.random.randn(100)
sigma = 5
u = 70
X = np.round(Xsn*sigma +u)
print(X)
```

```
plt.hist(X)
plt.show()
```

```
plt.style.use("dark_background")
```

```
plt.hist(X)
plt.show()
```

```
plt.hist(X)
plt.ylabel("Probability /Frequency count of students")
plt.xlabel("Students Marks Range")
plt.title("Histogram")
plt.show()
```

```
X2 = np.round(Xsn*5+40)
print(X)
print(X2)
```

```
plt.hist(X)
plt.hist(X2)
plt.show()
```

```
plt.hist(X, alpha = 0.8)
plt.hist(X2, alpha = 0.8)
plt.show()
```

```
plt.hist(X, alpha = 0.8, label = "Machine Learning", color = "red")
plt.hist(X2, alpha = 0.8, label = "Graph Theory", color = "green")
plt.legend()
plt.show()
```
