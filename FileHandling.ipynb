## Smarter way to open files
  Using with no need to close the file all the time!

```
with open ("textfile2.txt","r") as file:
  print (file.read())
  file.seek(5)
  print(file.read())
```

# Working with JSON
    Javascript Object Notation(JSON) - We need to share data among different programming languages.
    JSON works like python Dictionary.
    

```
with open("data.json","r") as file:
  print(file.read())
```

```
import json
with open("data.json","r") as file:
  d=json.load(file)
  print(d)
  print(type(d))
```

```
print(d["Name"])
```

```
print(d["Branch"])
```

```
def div(a,b):
    try :
      print(a/b)
    except ZeroDivisionError :
      print("Division by Zero is not allowed")
div(21,0)
```
