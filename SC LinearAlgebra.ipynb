**Detreminant**

```
A = np.array([[1, 2],
              [3, 4]])
np.linalg.det(A)
```

**Inverse**

```
Ainv = np.linalg.inv(A)
print(Ainv)   # Inverse
print(np.dot(A, Ainv)) # A*Ainv = Identity Matrix
```

**Pseudo** **Inverse**
- Moore - Penrose pseudo inverse is a generalization of the matrix inverse when the matrix may not be invertable.
- When a matrix is non-invertable i.e. 
                                        Det = 0 ==> Inverse  != Pseudo Inverse
- When a matrix is invertable i.e.                                         
                                        Det != 0 ==> Inverse  = Pseudo Inverse

```
Pinv = np.linalg.pinv(A)
print(Pinv)
print(np.dot(A, Pinv))   # Pseudo Inverse
```

```
B = np.array([[6, 8],
              [3, 4]])
np.linalg.det(B)
#Binv = np.linalg.inv(B)
#print(Binv)    # Inverse
#print(np.dot(B, Binv)) # A*Ainv = Identity Matrix

BPinv = np.linalg.pinv(B)
print(BPinv)     # Pseudo Inverse
print(np.dot(B, BPinv))
```

```
Binv = np.linalg.inv(B)
print(Binv)    # Inverse
print(np.dot(B, Binv)) # A*Ainv = Identity Matrix
```
