## **Probability Distribution & Statistics**

## Generate Standard Normal Distribution
   u =0, sigma =1

```
import numpy as np
import matplotlib.pyplot as plt
vals = np.random.randn(100)
vals = np.round(vals)
print(vals.shape)
print(vals)
```

```
plt.hist(vals, 10)
plt.show()
```

```
plt.hist(vals, 100)
plt.show()
```

## Generate Normal Distribution
   u !=0, sigma !=1

```
u = 5 
sigma = 2
vals = u + sigma * np.random.randn(100)
vals = np.round(vals)
print(vals.shape)
print(vals)
```

```
plt.hist(vals, 50)
plt.show()
```

```
u = 5 
sigma = 1
vals = u + sigma * np.random.randn(100)
vals = np.round(vals)
plt.hist(vals, 100)
plt.show()
```

```
# The 1st array represents the array
# The 2nd array represents the frequency count of the number
z = np.unique(vals, return_counts=True)
print(z)
```

```
# Bivariate 
mean = np.array([0.0, 0.0])
cov = np.array([[1, 0], [0, 1]])
dist = np.random.multivariate_normal(mean, cov, 500)
#print(dist)
print(dist.shape)
```

```
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```
