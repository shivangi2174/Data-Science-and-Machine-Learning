# Face Recognition

```
import cv2
from matplotlib import pyplot as plt
```

### ***Read an image***

```
img = cv2.imread('/content/drive/MyDrive/Python/teddy.png')
```

```
plt.imshow(img)
plt.show()
```

### **Convertion between color spaces**

```
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
```
